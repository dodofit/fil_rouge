```python
#2022
df_preds = pd.read_csv('/Users/dorianfitton/Documents/Cours_T칠l칠com/fil_rouge/fil_rouge/examples/experimental_files/fc_nl_datFR_new_YT1_CW1456_2.csv')
len_ = len(df_preds)
MAE_list = []
sMAPE_list = []
df_preds=df_preds.dropna()
for i in range(len_):
    date = df_preds.iloc[i]['Date']
    # Saving the current prediction
    Yp = df_preds.iloc[i].drop('Date')
    real_values = df_values['Prices'].loc[pd.to_datetime(date+' 00:00:00'):pd.to_datetime(date+' 23:00:00')]

    # Computing metrics up-to-current-date
    mae = np.mean(MAE(Yp.values.squeeze(), real_values.values))
    smape = np.mean(sMAPE(Yp.values.squeeze(), real_values.values)) * 100
    MAE_list.append(mae)
    sMAPE_list.append(smape)

    # Pringint information
    print('{} - sMAPE: {:.2f}%  |  MAE: {:.3f}'.format(str(date)[:10], smape, mae))
```
